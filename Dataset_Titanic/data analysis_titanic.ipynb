```python
#observando os 10 valores das passagens mais barata
df.sort_values('Fare').head(10)
```

```python
#observando os 10 valores de passagens mais caros
df.sort_values("Fare", ascending = False).head(10)
```

```python
#salvando os dados classificando-os pelas passagens mais caras
df.sort_values("Fare", ascending = False, inplace = True)
```

```python
#observando as modificações
df.head()
```

```python
#tratando dados faltantes
df.sort_values("Fare", ascending = False, na_position='last').tail(20) #.tail nos mostra os ultimos 20 elementos do conjunto de dados
```

```python
#contando a ocorrência de uma variavel em determinada coluna
df["Sex"].value_counts()
```

```python
#verificando o numero de valores exclusivos que ocorrem no conjunto de dados
df.nunique()
```

```python
#mudando o tipo de dado
df["Embarked"] = df["Embarked"].astype("category")
```

```python
df["Embarked"].dtype
```

```python
#filtrando com uma condição
df[df["Embarked"] == 'C']
```

```python
#filtrando com duas ou mais condições usando o operador and
df_fare_mask = df["Fare"] < 100
df_sex_mask = df["Sex"] == "female"
df[df_fare_mask & df_sex_mask]
```

```python
#filtrando com duas ou mais condições usando o operador or
df_fare_mask2 = df["Fare"] > 500
df_age_mask = df["Age"] > 70
df[df_fare_mask2 | df_age_mask]
```

```python
#encontrando valores null em uma coluna
null_mask = df["Cabin"].isnull()
df[null_mask]
```

```python
#observando os valores nulos em todo conjunto de dados
df.isnull().sum()
```

```python
#lidando com valores ausentes
df.drop(labels = ["Cabin"], axis = 1).head()
```

```python
#preenchendo valores ausentes
df['Age'] = df['Age'].fillna((df['Age'].median()))
```
